## 2. Data Cleaning
We handle missing values, duplicates, and data type issues to prepare the dataset for analysis.

```python
import logging
import os


logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)


input_file = 'SA_Aqar.csv'  
try:
    logger.info(f"Loading dataset from {input_file}")
    df = pd.read_csv(input_file, low_memory=False) 
except FileNotFoundError:
    logger.error(f"File {input_file} not found. Please check the file path.")
    raise
except Exception as e:
    logger.error(f"Error loading dataset: {str(e)}")
    raise

logger.info(f"Dataset shape: {df.shape}")
logger.info(f"Columns: {list(df.columns)}")

logger.info("Checking for missing values")
print("Missing Values:\n", df.isnull().sum())


if 'price' in df.columns:
    logger.info("Filling missing 'price' values with median")
    df['price'] = df['price'].fillna(df['price'].median())
else:
    logger.warning("'price' column not found")
if 'city' in df.columns:
    logger.info("Filling missing 'city' values with mode")
    mode_city = df['city'].mode()
    df['city'] = df['city'].fillna(mode_city[0] if not mode_city.empty else 'Unknown')
else:
    logger.warning("'city' column not found")

required_columns = ['price', 'size']
if all(col in df.columns for col in required_columns):
    logger.info("Dropping rows with missing 'price' or 'size'")
    df = df.dropna(subset=required_columns)
else:
    missing_cols = [col for col in required_columns if col not in df.columns]
    logger.error(f"Missing required columns: {missing_cols}")
    raise ValueError(f"Missing required columns: {missing_cols}")


logger.info("Removing duplicate rows")
df = df.drop_duplicates()


if 'price' in df.columns:
    logger.info("Converting 'price' to numeric")
    df['price'] = pd.to_numeric(df['price'], errors='coerce')
if 'size' in df.columns:
    logger.info("Converting 'size' to numeric")
    df['size'] = pd.to_numeric(df['size'], errors='coerce')
if 'price' in df.columns:
    logger.info("Handling outliers in 'price'")
    q1 = df['price'].quantile(0.25)
    q3 = df['price'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    logger.info(f"Price bounds: [{lower_bound}, {upper_bound}]")
    df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
else:
    logger.warning("Cannot handle outliers; 'price' column missing")

output_dir = 'data/processed'
output_file = f'{output_dir}/cleaned_data.csv'
try:
    logger.info(f"Saving cleaned data to {output_file}")
    os.makedirs(output_dir, exist_ok=True)
    df.to_csv(output_file, index=False)
    logger.info(f"Cleaned data saved successfully. Final shape: {df.shape}")
except Exception as e:
    logger.error(f"Error saving cleaned data: {str(e)}")
    raise

print("Data cleaning completed successfully!")
```

## 3. Exploratory Data Analysis
### 3.1 Distribution of Property Prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='blue')
plt.title('Distribution of Property Prices in Saudi Arabia')
plt.xlabel('Price (SAR)')
plt.ylabel('Frequency')
plt.show()
```

**Insight**: The price distribution is [e.g., right-skewed, with most properties priced below X SAR]. There are a few high-value properties, indicating a diverse market.

### 3.2 Average Propertys Prices

```python
df['price'].plot(kind='box')
```

### 3.3 Average Prices By Each City

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='city', y='price', data=df)
plt.title('Property Prices by City')
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Price (SAR)')
plt.show()
```

**Insight**: [e.g., Riyadh has the highest median prices, while Dammam is more affordable. Jeddah shows high variability, suggesting a mix of luxury and budget properties.]

### 3.3 Correlation Analysis

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
#plt.savefig('assets/figures/correlation_matrix.png')
plt.show()
```

**Insight**: [e.g., Price strongly correlates with size (r=0.8) and number of bedrooms (r=0.6), indicating these are key drivers of value.]


**Insight**: [e.g., Apartments dominate in Riyadh, while villas are more common in Jeddah, reflecting urban vs. suburban preferences.]

```python
def categorize_age(age):
    if pd.isna(age) or age < 0:
        return 'Unknown' 
    if age < 5:
        return 'New'
    elif age <= 15:
        return 'Medium'
    else:
        return 'Old'


if 'property_age' in df.columns:
    df['age_category'] = df['property_age'].apply(categorize_age)
else:
    print("Error: 'property_age' column not found in the dataset.")
```
